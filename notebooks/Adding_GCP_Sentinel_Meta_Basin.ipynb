```python
sent_idx = pd.read_csv("index.csv")
```

```python
def query_df(meta_lat, meta_long, idx_df):
  floor_lat = math.floor(meta_lat)
  floor_long = math.floor(meta_long)
  ceil_lat = math.ceil(meta_lat)
  ceil_long = math.ceil(meta_long)
  idx_lat_filt = idx_df.query("@floor_lat<=NORTH_LAT<=@ceil_lat")
  idx_lon_filt = idx_lat_filt.query("@floor_long<=WEST_LON<=@ceil_long")
  return idx_lon_filt
```
